# Problem 8

```python
year = df.year - pd.DatetimeIndex(df.birth_date).year
oldest = year.argmax()
youngest = year.argmin()
print("oldest is", df['full_name'][oldest], "and is" ,year.max())
print("youngest is", df['full_name'][youngest], "and is" ,year.min())
print("average age", year.mean())
print("standard deviation", year.std())
```

```python
fig = plt.figure()
axes = fig.add_axes([0.2,0.2,1.8,1.8])
axes.hist(year);
axes.set_xlabel('Age')
axes.set_ylabel('Winners');
```
