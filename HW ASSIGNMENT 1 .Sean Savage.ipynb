```python
"Problem 1"
import math
import numpy as np
import matplotlib.pyplot as plt
n = np.arange(2,20)
stirling = n * np.log(n) - n
exact = np.zeros(len(n))
for i in range(len(n)):
    fact = np.math.factorial(n[i])
    exact[i] = np.log(fact)
print(stirling)
print(exact)
```

```python
fig = plt.figure()
import matplotlib.pyplot as plt
%matplotlib notebook
plt.plot(exact,'r-',label='exact')
plt.plot(stirling,'b-',label='stirling')
plt.xlabel('n value', fontsize=18)
plt.ylabel('output value', fontsize=16)
plt.legend()
```

```python
 "Problem 1 with error"
n = np.arange(2,91, dtype = float)
stirling = n * np.log(n) - n
exact = np.zeros(len(n))
for i in range(len(n)):
    fact = float(np.math.factorial(n[i]))
    exact[i] = float(np.log(fact))
print(stirling)
print(exact)
error = ((stirling - exact)/exact) * 100
#if error =< 0.01:
print(error)
#print(error[error<0.1])
print("the error is less then 1% at 91")
```

```python
"Problem 5a"
np.random.seed(0)
sd= np.random.randn(1000,1000)
x5 = sd[0]
y5 = sd[:,0]
print(sd)
```

```python
"Problem 5b"
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1,2)

axes[0].hist(x5,bins=20)
axes[1].hist(y5,bins=20)
axes[0].set_title('random x value')
axes[0].set_xlabel('x')
axes[0].set_ylabel('frequency')
axes[1].set_title('random y value')
axes[1].set_xlabel('y')
axes[1].set_ylabel('frequency')
plt.tight_layout()

#x variable with 1000
#y variable with 1000 variables
```
